```python
import pandas as pd
import wandb

wandb.login()
```

```python
api = wandb.Api()

runs = api.runs("jeroenbe/structure-learning")

total_list = []
for run in runs:
    total_list.append({'name': run.name, **dict(run.summary._json_dict),**dict(run.config.items())})
    
run_df = pd.DataFrame(total_list)
```

```python
group_df = run_df.groupby(['d', 'n', 'model', 'graph_type', 's0', 'sem_type'])
```

```python
group_df.mean()
```

```python
group_df.count()
```

```python
for name, group in group_df:
    print(f"{name}")
    
    print((f"{round(group.shd_normalised.mean(), 3)} \\scriptsize{{({round(group.shd_normalised.std(), 3)})}} &"
           f"{round(group.shd.mean(), 3)} \\scriptsize{{({round(group.shd.std(), 3)})}} & "
           f"{round(group.fpr_normalised.mean(), 3)} \\scriptsize{{({round(group.fpr_normalised.std(), 3)})}} & "
           f"{round(group.fpr.mean(), 3)} \\scriptsize{{({round(group.fpr.std(), 3)})}} & "
           f"{round(group.tpr_normalised.mean(), 3)} \\scriptsize{{({round(group.tpr_normalised.std(), 3)})}} & "
           f"{round(group.tpr.mean(), 3)} \\scriptsize{{({round(group.tpr.std(), 3)})}} & "
           f"{round(group.fdr_normalised.mean(), 3)} \\scriptsize{{({round(group.fdr_normalised.std(), 3)})}} & "
           f"{round(group.fdr.mean(), 3)} \\scriptsize{{({round(group.fdr.std(), 3)})}} & "
           f"{round(group['DAG-diff'].mean(), 3)} \\scriptsize{{({round(group['DAG-diff'].std(), 3)})}}"))
    print("--------\n")
```
